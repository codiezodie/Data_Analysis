```python
import numpy as np
from scipy import stats 
```

```python
array_3d = np.random.randint(1, 100, (3, 4, 5))  # Random integers from 1 to 99
print("3D Array:\n", array_3d)
```

```python
#Calculating Max
max_value = np.max(array_3d)
print("Maximum value:", max_value)
```

```python
#Calculating Min
min_value = np.min(array_3d)
print("Minimum value:", min_value)
```

```python
#Mean 
mean_value = np.mean(array_3d)
print("Mean value:", mean_value)
```

```python
#Mode
mode_value = stats.mode(array_3d, axis=None)
print("Mode value:", mode_value.mode.item(), " | Count:", mode_value.count.item())

```

```python
#Median
median_value = np.median(array_3d)
print("Median value:", median_value)
```
