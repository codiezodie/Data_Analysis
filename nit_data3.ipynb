```python
Lang = df['LanguageWorkedWith']
Lang
```

```python
Lang_counter = Counter()
for response in Lang:
    response = str(response)
    Lang_counter.update(response.split(;))

```

```python
languages = []
most_used = []

for item in Lang_counter.most_common(15):
    languages.append(item[0])
    most_used.append(item[1])

plt.bar(languages, most_used)
plt.title("Most Populer Language")
plt.ylabel("Programming Language")
plt.xlabel("Number of People Who use")
```
